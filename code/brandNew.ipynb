```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
dataset1 = 'data/AirVisual Data Dump.csv'
datasetAirPort = 'data/airportdata.csv'
```

```python
data = pd.read_csv(dataset1)
dataPort = pd.read_csv(datasetAirPort)
dataPort.info()
data.info()
```

```python
dataPort.groupby(["station"]).count()
```

```python
display(data.shape)
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
dataPort.drop(["gust_mph", "peak_wind_gust_mph"], inplace = True, axis = 1)
dataPort.head()
```

```python
data["City"].fillna("Karachi", inplace = True)
```

```python
dataPort.info()
data.info()
```

```python
dataPort['Datetime'] = pd.to_datetime(dataPort.valid)
dataPort['Datetime'] = dataPort['Datetime'].dt.round('H')
dataPort.groupby("Datetime").count()
```

```python
# OPFA Faisalabad
# OPKC Karachi
# OPLA Lahore
# OPPS Peshawar
# OPRN Islamabad
serialToCity = {'OPFA': 'Faisalabad', 'OPKC': 'Karachi', 'OPLA': 'Lahore', 'OPPS': 'Peshawar', 'OPRN': 'Islamabad'}
```

```python
dataPort['City'] = dataPort.station.apply(lambda x : serialToCity[x])
dataPort.drop(['valid','station', 'lon', 'lat'], axis = 1, inplace=True)
dataPort.drop(dataPort[dataPort.City=="Faisalabad"].index, inplace=True)
```

```python
data.shape
```

```python
display(dataPort.head(100))
dataPort = dataPort.groupby(['City', 'Datetime']).mean()
```

```python
data.Datetime = pd.to_datetime(data['Datetime'])
data = data.merge( dataPort, on = ['Datetime', 'City'])
data.drop(["latitude", "longitude", "Outdoor PM2.5", "OutdoorUSAQI"], axis = 1, inplace = True)
data.to_csv('data/mergedData.csv')
```

```python
data['weekday'] = data['Datetime'].dt.dayofweek
data['hour'] = data.Datetime.dt.hour
data['month'] = data.Datetime.dt.month
data['year'] = data.Datetime.dt.year
data['dayofmonth'] = data.Datetime.dt.day
```
